```python
import tensorflow as tf
import matplotlib.pyplot as plt

# Load the image
image_path = "COLOURS.png"  # Make sure the file name matches
image = tf.io.read_file(image_path)
image = tf.io.decode_png(image, channels=1)  # Convert to grayscale
image = tf.image.resize(image, size=[300, 300])

# Display the image
plt.figure(figsize=(5, 5))
plt.imshow(tf.squeeze(image), cmap="gray")
plt.axis("off")
plt.title("Original Image")
plt.show()
```

```python
# Define the kernel
kernel = tf.constant([[-1, -1, -1],
                      [-1,  8, -1],
                      [-1, -1, -1]], dtype=tf.float32)

# Convert image format for TensorFlow
image = tf.image.convert_image_dtype(image, dtype=tf.float32)
image = tf.expand_dims(image, axis=0)  # Add batch dimension
kernel = tf.reshape(kernel, [*kernel.shape, 1, 1])

# Apply Convolution
conv_fn = tf.nn.conv2d
image_filter = conv_fn(input=image, filters=kernel, strides=1, padding='SAME')

# Apply Activation Function (ReLU)
relu_fn = tf.nn.relu
image_detect = relu_fn(image_filter)

# Apply Pooling Layer
pool = tf.nn.pool
image_condense = pool(input=image_detect, window_shape=(2, 2), pooling_type='MAX', strides=(2, 2), padding='SAME')

# Display Results
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(tf.squeeze(image_filter))
plt.axis("off")
plt.title("Convolution")

plt.subplot(1, 3, 2)
plt.imshow(tf.squeeze(image_detect))
plt.axis("off")
plt.title("Activation")

plt.subplot(1, 3, 3)
plt.imshow(tf.squeeze(image_condense))
plt.axis("off")
plt.title("Pooling")

plt.show()
```
